```
import sympy as sp

n = sp.Symbol("n", positive=True)

print("Respuesta Ejercicio 1")

an = (6*n**3 - 4*n**2 + 7)/(3*n**3 + 2*n**2 - 5*n + 1)

sp.pprint(sp.limit(an, n, sp.oo))
```

```
import sympy as sp

n = sp.Symbol("n", positive=True)

print("Respuesta Ejercicio 2")

a = (5*n**2 + 3*n + 1)/(2*n**4 + n**2 + 4)
b = 1/n**2

sp.pprint(sp.limit(a/b, n, sp.oo))
```

```
import sympy as sp

x = sp.Symbol("x")
y = sp.Function("y")

print("Respuesta Ejercicio 3")

ecuacion = sp.Eq(
    sp.diff(y(x), x),
    3*x**2*(y(x)-2)
)

respuesta = sp.dsolve(ecuacion)

sp.pprint(respuesta)
```

```
import sympy as sp

x = sp.Symbol("x")
v = sp.Function("v")

print("Respuesta Ejercicio 4")

y = x*v(x)

ecuacion = sp.Eq(
    sp.diff(y, x),
    (2*x + y)/(x - y)
)

sp.pprint(sp.simplify(ecuacion))
```

```
import sympy as sp

t, s = sp.symbols("t s")

print("Respuesta Ejercicio 10")

f = t**3*sp.exp(-t) + sp.exp(2*t)*sp.cos(5*t) + 4*sp.sin(3*t)

respuesta = sp.laplace_transform(f, t, s, noconds=True)

sp.pprint(respuesta)
```

```
import sympy as sp

t, s = sp.symbols("t s")
y = sp.Function("y")

print("Respuesta Ejercicio 11")

ecuacion = sp.Eq(
    sp.diff(y(t), t) + 4*y(t),
    t**2*sp.exp(-t) + sp.sin(2*t)
)

respuesta = sp.dsolve(
    ecuacion,
    ics={y(0): 2}
)

sp.pprint(respuesta)
```

```
import sympy as sp

N = 6

x = sp.Symbol("x")
a = sp.symbols(f"a0:{N+1}")

print("Respuesta Ejercicio 12")

y = sum(a[n]*x**n for n in range(N+1))

y1 = sp.diff(y, x)
y2 = sp.diff(y, x, 2)

ecuacion = sp.expand(y2 + x*y1 - 2*y)

sp.pprint(ecuacion)
```

```
import sympy as sp

N = 6

x = sp.Symbol("x")
a = sp.symbols(f"a0:{N+1}")

print("Respuesta Ejercicio 13")

y = sum(a[n]*x**n for n in range(N+1))

y1 = sp.diff(y, x)
y2 = sp.diff(y, x, 2)

ecuacion = sp.expand(y2 - x*y1 + 2*y)

condiciones = [
    sp.Eq(y.subs(x,0),2),
    sp.Eq(y1.subs(x,0),1)
]

sp.pprint(ecuacion)

for c in condiciones:
    sp.pprint(c)
```

```
import sympy as sp

t = sp.Symbol("t")

x = sp.Function("x")
y = sp.Function("y")

print("Respuesta Ejercicio 14")

ec1 = sp.Eq(
    sp.diff(x(t),t),
    3*x(t)+2*y(t)
)

ec2 = sp.Eq(
    sp.diff(y(t),t),
    x(t)+4*y(t)
)

respuesta = sp.dsolve((ec1,ec2))

for r in respuesta:
    sp.pprint(r)
```

```
import sympy as sp

t = sp.Symbol("t")
lamda = sp.Symbol("lamda")

x = sp.Function("x")
y = sp.Function("y")

print("Respuesta Ejercicio 15")

A = sp.Matrix([
    [5,2],
    [2,2]
])

sp.pprint(A)

print("Determinante")

sp.pprint((A-lamda*sp.eye(2)).det())

print("Raices")

sp.pprint(sp.solve((A-lamda*sp.eye(2)).det(),lamda))

ec1 = sp.Eq(
    sp.diff(x(t),t),
    5*x(t)+2*y(t)
)

ec2 = sp.Eq(
    sp.diff(y(t),t),
    2*x(t)+2*y(t)
)

respuesta = sp.dsolve((ec1,ec2))

for r in respuesta:
    sp.pprint(r)
```
